# Seaborn – Class 4: Regression & Mixed Plots

## Regression & Mixed Plots

Seaborn provides plots that combine **relationships** and **statistical models** or mix multiple plot types.

- `regplot()` / `lmplot()` → scatter + regression line  
- `jointplot()` → scatter with hist/KDE  
- `pairplot()` → multiple relationships in one go


# regplot() – Scatter + Regression Line

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')
tips
```

```python
# regplot()
sns.regplot(x=tips['total_bill'],y = tips['tip'],color='red')
```

# lmplot() – Regression with Facets

```python
sns.lmplot(x= 'total_bill', y = 'tip',hue='sex', data=tips, height=5, aspect=1.2)
```

# jointplot() – Scatter + Histogram / KDE

```python
sns.jointplot(x=tips['total_bill'],y=tips['tip'],kind='scatter')
```

```python
sns.jointplot(x=tips['total_bill'],y=tips['tip'],kind='hex',color='red')
```

```python
sns.jointplot(x=tips['total_bill'],y=tips['tip'],kind='kde',fill='true',color='green') #kind='scatter'or'hex'or'kde'
```

# pairplot() – Relationships Between Many Variables

```python
iris = sns.load_dataset('iris')
iris
```

```python
sns.pairplot(iris,hue='species')
```
